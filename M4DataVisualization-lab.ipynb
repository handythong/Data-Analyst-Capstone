## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT ConvertedComp
FROM master
"""
df = pd.read_sql_query(QUERY,conn)

count, bin_edges = np.histogram(df.dropna())

df.dropna().plot(kind='hist', xticks=bin_edges)
plt.title('Histogram of Converted Comp')
plt.xlabel('Converted Comp')
plt.ylabel('Number of people')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age
FROM master
"""
df1 = pd.read_sql_query(QUERY,conn)

df1.plot(kind='box', color='blue', grid=True)
plt.title('Boxplot of Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
df2 = pd.read_sql_query(QUERY,conn)
df2.plot(kind='scatter', x='Age', y='WorkWeekHrs')
plt.title('Scatter plot of Age and Work Week Hours')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""
df3 = pd.read_sql_query(QUERY,conn)
df3 = df3.dropna()
norm_age = (df3['Age']-df3['Age'].min()) / (df3['Age'].max()-df3['Age'].min())
df3.plot(kind='scatter', x='WorkWeekHrs', y='CodeRevHrs', s=norm_age*200)
plt.title('Bubble plot of Work Week Hours and Code Rev Hours')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT DatabaseDesireNextYear, count(DatabaseDesireNextYear) as Count
FROM DatabaseDesireNextYear
group by DatabaseDesireNextYear
order by Count desc limit 5
"""
df4 = pd.read_sql_query(QUERY,conn)
lab = df4['DatabaseDesireNextYear']
explode_list=[0,0,0.1,0.1,0.1]
df4.plot(kind='pie', y='Count', labels=lab, autopct='%1.1f%%', figsize=(5,5), startangle=90, shadow=True, explode=explode_list, legend=None)
plt.title('Top 5 Databases Respondent Wish to Learn Next Year')
plt.show()
```

```python
QUERY = """
SELECT LanguageDesireNextYear, count(LanguageDesireNextYear) as Count
FROM LanguageDesireNextYear
group by LanguageDesireNextYear
order by Count desc limit 5
"""
df5 = pd.read_sql_query(QUERY,conn)
df5
```

```python
QUERY = """
SELECT LanguageWorkedWith, count(LanguageWorkedWith) as Count
FROM LanguageWorkedWith
group by LanguageWorkedWith
order by Count desc limit 5
"""
df6 = pd.read_sql_query(QUERY,conn)
df6
```

```python
QUERY = """
SELECT *
FROM DatabaseWorkedWith
"""
df7 = pd.read_sql_query(QUERY,conn)
df7 = df7.drop_duplicates('Respondent', keep=False)
df7['DatabaseWorkedWith'].value_counts()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
Where Age > 30 AND Age < 36
"""
df8 = pd.read_sql_query(QUERY,conn)
df8 = df8.groupby(by='Age').median()
df8.plot(kind='bar', stacked=True)
plt.title('Stacked Bar Charts for Median of Work Week Hours and Code Rev Hours According to Age Group')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT Age, ConvertedComp
FROM master
Where Age > 45 AND Age < 61
"""
df9 = pd.read_sql_query(QUERY,conn)
df9 = df9.groupby(by='Age').median()
df9.plot(kind='line')
plt.title('Line Plot for Median of Converted Comp According to Age Group')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch
FROM master
"""
df10 = pd.read_sql_query(QUERY,conn)
df10.value_counts().plot(kind='barh')
plt.title('Bar Chart of Main Branch')
plt.show()
```
